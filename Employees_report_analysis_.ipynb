```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data =pd.read_csv('/content/Employees.csv')
data.head()
```

```python
data.info()
# no missing values
```

```python
data.describe()
```

```python

data['Gender'].value_counts().plot(kind='pie', autopct = "%1.0f%%")
```

```python
# number of employees in each country

import matplotlib.pyplot as plt
sns.countplot(y='Country', data=data)
plt.show()
```

```python
# number of employees in each departement
sns.countplot(y='Department', data=data)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Calculate total salaries for each country
country_salaries = data.groupby('Country')['Monthly Salary'].sum().sort_values(ascending=False)
country_salaries
```

```python
# show monthly salaries for each country


plt.figure(figsize=(10, 6))
sns.barplot(x=country_salaries.values, y=country_salaries.index)
plt.xlabel('Total Salary')
plt.ylabel('Country')
plt.title('Total Salaries by Country')
plt.show()
```

```python
#  plot number of employees by center

import matplotlib.pyplot as plt
sns.countplot(y='Center', data=data)
plt.show()
```

```python
# prompt: averrage job rate for each country

# Calculate average job satisfaction rate for each country
average_job_rate = data.groupby('Country')['Job Rate'].mean().sort_values(ascending=False)

# Display the result
average_job_rate
```
